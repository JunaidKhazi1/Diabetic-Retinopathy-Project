```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset, WeightedRandomSampler
from torchvision import transforms, models, datasets
import numpy as np
from PIL import Image
import os
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import time
import optuna
from collections import Counter
import cv2

# GLCM features
from skimage.feature import graycomatrix, graycoprops
from skimage import measure
from skimage.feature import local_binary_pattern

# Metrics
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

import warnings
warnings.filterwarnings('ignore')

# Set device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"ğŸš€ Using device: {device}")

class GLCMFeatureExtractor:
    """GLCM Feature Extractor optimized for DR classification"""
    
    def __init__(self, distances=[1, 2], angles=[0, 45, 90, 135], levels=32):
        self.distances = distances
        self.angles = [np.deg2rad(angle) for angle in angles]
        self.levels = levels
        print(f"ğŸ”¬ GLCM Extractor: {len(distances)} distances, {len(angles)} angles, {levels} levels")
    
    def extract_glcm_features(self, image):
        """Extract compact GLCM features from PIL image"""
        try:
            # Convert PIL to numpy and preprocess
            if isinstance(image, Image.Image):
                img_array = np.array(image)
            else:
                img_array = image
            
            # Use green channel for better retinal contrast
            if len(img_array.shape) == 3:
                gray = img_array[:, :, 1]  # Green channel
            else:
                gray = img_array
            
            # Resize and enhance
            gray = cv2.resize(gray, (128, 128))  # Smaller for speed
            gray = cv2.equalizeHist(gray)
            
            # Reduce gray levels
            gray = (gray // (256 // self.levels)).astype(np.uint8)
            
            # Extract GLCM features
            features = []
            
            for distance in self.distances:
                glcm = graycomatrix(gray, distances=[distance], angles=self.angles, 
                                 levels=self.levels, symmetric=True, normed=True)
                
                # Core texture properties
                contrast = graycoprops(glcm, 'contrast').flatten()
                homogeneity = graycoprops(glcm, 'homogeneity').flatten()
                energy = graycoprops(glcm, 'energy').flatten()
                
                # Add mean and std for each property
                features.extend([
                    np.mean(contrast), np.std(contrast),
                    np.mean(homogeneity), np.std(homogeneity),
                    np.mean(energy), np.std(energy)
                ])
            
            return np.array(features, dtype=np.float32)
            
        except Exception as e:
            print(f"âŒ GLCM extraction error: {e}")
            return np.zeros(len(self.distances) * 6, dtype=np.float32)  # 6 features per distance

class FolderDRDataset(Dataset):
    """Dataset for folder-based DR classification with GLCM features"""
    
    def __init__(self, root_dir, transform=None, extract_glcm=True):
        self.root_dir = root_dir
        self.transform = transform
        self.extract_glcm = extract_glcm
        
        if self.extract_glcm:
            self.glcm_extractor = GLCMFeatureExtractor()
        
        # Load samples from folders
        self.samples = []
        self.classes = []
        self.class_to_idx = {}
        
        print(f"ğŸ“ Loading dataset from: {root_dir}")
        
        # Get class folders
        for class_name in sorted(os.listdir(root_dir)):
            class_path = os.path.join(root_dir, class_name)
            if os.path.isdir(class_path):
                self.classes.append(class_name)
                class_idx = len(self.classes) - 1
                self.class_to_idx[class_name] = class_idx
                
                # Load images from this class folder
                image_count = 0
                for img_name in os.listdir(class_path):
                    if img_name.lower().endswith(('.png', '.jpg', '.jpeg')):
                        img_path = os.path.join(class_path, img_name)
                        self.samples.append((img_path, class_idx))
                        image_count += 1
                
                print(f"   ğŸ“Š Class {class_name} (ID: {class_idx}): {image_count} images")
        
        print(f"âœ… Loaded {len(self.samples)} total samples from {len(self.classes)} classes")
        
        # Calculate class distribution
        labels = [label for _, label in self.samples]
        self.class_counts = Counter(labels)
        
        # Calculate sample weights for balanced sampling
        total_samples = len(self.samples)
        self.class_weights = {cls_idx: total_samples / count 
                             for cls_idx, count in self.class_counts.items()}
        self.sample_weights = [self.class_weights[label] for _, label in self.samples]
    
    def __len__(self):
        return len(self.samples)
    
    def __getitem__(self, idx):
        img_path, label = self.samples[idx]
        
        # Load image
        try:
            image = Image.open(img_path).convert('RGB')
        except Exception as e:
            print(f"âŒ Error loading {img_path}: {e}")
            image = Image.new('RGB', (224, 224), (0, 0, 0))
        
        # Extract GLCM features before transform
        if self.extract_glcm:
            glcm_features = self.glcm_extractor.extract_glcm_features(image)
        else:
            glcm_features = torch.zeros(12, dtype=torch.float32)  # Default size
        
        # Apply transforms
        if self.transform:
            image = self.transform(image)
        
        return image, torch.tensor(glcm_features, dtype=torch.float32), label

class CombinedDenseNetClassifier(nn.Module):
    """DenseNet121 + GLCM Combined Classifier"""
    
    def __init__(self, n_classes, glcm_size=12):
        super(CombinedDenseNetClassifier, self).__init__()
        print(f"ğŸ—ï¸  Building DenseNet121 + GLCM model...")
        
        # DenseNet121 backbone
        self.base_model = models.densenet121(pretrained=True)
        self.base_model.classifier = nn.Identity()
        self.feature_size = 1024
        self.glcm_size = glcm_size
        
        # Combined classifier
        self.classifier = nn.Sequential(
            nn.BatchNorm1d(self.feature_size + self.glcm_size, momentum=0.999, eps=0.001),
            nn.Linear(self.feature_size + self.glcm_size, 1024),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(1024, 512),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(512, 256),
            nn.ReLU(),
            nn.Dropout(0.4),
            nn.Linear(256, n_classes)
        )
        
        print(f"âœ… Model built: {self.feature_size} CNN + {self.glcm_size} GLCM â†’ {n_classes} classes")
    
    def forward(self, x, glcm_feats):
        # DenseNet features
        x = self.base_model(x)
        
        # Combine CNN and GLCM features
        combined = torch.cat((x, glcm_feats), dim=1)
        
        # Final classification
        return self.classifier(combined)

def create_data_loaders(data_dir, batch_size=16, val_split=0.2, test_split=0.1):
    """Create train/val/test data loaders from folder structure"""
    
    print("ğŸ¨ Setting up data transformations...")
    
    # Data transformations
    train_transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.RandomHorizontalFlip(p=0.5),
        transforms.RandomRotation(degrees=10),
        transforms.ColorJitter(brightness=0.2, contrast=0.2),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])
    
    val_test_transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])
    
    # Create full dataset
    print("ğŸ“¦ Creating full dataset...")
    full_dataset = FolderDRDataset(data_dir, transform=None, extract_glcm=True)
    
    # Create stratified splits
    print("âœ‚ï¸  Creating stratified train/val/test splits...")
    
    # Group samples by class
    class_samples = {i: [] for i in range(len(full_dataset.classes))}
    for idx, (_, label) in enumerate(full_dataset.samples):
        class_samples[label].append(idx)
    
    train_indices, val_indices, test_indices = [], [], []
    
    for class_id, samples in class_samples.items():
        np.random.shuffle(samples)
        n_samples = len(samples)
        
        # Calculate split sizes
        test_size = max(1, int(test_split * n_samples))
        val_size = max(1, int(val_split * n_samples))
        train_size = n_samples - test_size - val_size
        
        # Split indices
        test_indices.extend(samples[:test_size])
        val_indices.extend(samples[test_size:test_size + val_size])
        train_indices.extend(samples[test_size + val_size:])
    
    # Shuffle indices
    np.random.shuffle(train_indices)
    np.random.shuffle(val_indices)
    np.random.shuffle(test_indices)
    
    print(f"ğŸ“Š Data split:")
    print(f"   Training:   {len(train_indices)} samples")
    print(f"   Validation: {len(val_indices)} samples") 
    print(f"   Test:       {len(test_indices)} samples")
    
    # Create subset datasets with transforms
    class TransformSubset(Dataset):
        def __init__(self, dataset, indices, transform):
            self.dataset = dataset
            self.indices = indices
            self.transform = transform
        
        def __len__(self):
            return len(self.indices)
        
        def __getitem__(self, idx):
            original_idx = self.indices[idx]
            img_path, label = self.dataset.samples[original_idx]
            
            # Load image
            try:
                image = Image.open(img_path).convert('RGB')
            except:
                image = Image.new('RGB', (224, 224), (0, 0, 0))
            
            # Extract GLCM features
            glcm_features = self.dataset.glcm_extractor.extract_glcm_features(image)
            
            # Apply transform
            if self.transform:
                image = self.transform(image)
            
            return image, torch.tensor(glcm_features, dtype=torch.float32), label
    
    # Create transformed datasets
    train_dataset = TransformSubset(full_dataset, train_indices, train_transform)
    val_dataset = TransformSubset(full_dataset, val_indices, val_test_transform)
    test_dataset = TransformSubset(full_dataset, test_indices, val_test_transform)
    
    # Create weighted sampler for training
    train_weights = [full_dataset.sample_weights[i] for i in train_indices]
    train_sampler = WeightedRandomSampler(train_weights, len(train_weights), replacement=True)
    
    # Create data loaders
    print("ğŸšš Creating data loaders...")
    train_loader = DataLoader(train_dataset, batch_size=batch_size, sampler=train_sampler,
                             num_workers=2, pin_memory=True)
    val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False,
                           num_workers=2, pin_memory=True)
    test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False,
                            num_workers=2, pin_memory=True)
    
    return train_loader, val_loader, test_loader, full_dataset.classes

def objective_densenet_glcm(trial, train_loader, val_loader, n_classes):
    """Optuna objective function for DenseNet + GLCM hyperparameter optimization"""
    
    # Suggest hyperparameters
    lr = trial.suggest_loguniform("lr", 1e-5, 1e-2)
    momentum = trial.suggest_uniform("momentum", 0.5, 0.99)
    batch_size = trial.suggest_categorical("batch_size", [8, 16, 32])
    dropout_1 = trial.suggest_uniform("dropout_1", 0.1, 0.4)
    dropout_2 = trial.suggest_uniform("dropout_2", 0.2, 0.5)
    dropout_3 = trial.suggest_uniform("dropout_3", 0.3, 0.6)
    
    print(f"ğŸ”§ Trial {trial.number}: lr={lr:.6f}, momentum={momentum:.3f}, batch_size={batch_size}")
    
    # Create model with trial-specific dropout
    class TrialCombinedClassifier(nn.Module):
        def __init__(self, n_classes, glcm_size=12):
            super(TrialCombinedClassifier, self).__init__()
            self.base_model = models.densenet121(pretrained=True)
            self.base_model.classifier = nn.Identity()
            self.feature_size = 1024
            self.glcm_size = glcm_size
            
            self.classifier = nn.Sequential(
                nn.BatchNorm1d(self.feature_size + self.glcm_size, momentum=0.999, eps=0.001),
                nn.Linear(self.feature_size + self.glcm_size, 1024),
                nn.ReLU(),
                nn.Dropout(dropout_1),
                nn.Linear(1024, 512),
                nn.ReLU(),
                nn.Dropout(dropout_2),
                nn.Linear(512, 256),
                nn.ReLU(),
                nn.Dropout(dropout_3),
                nn.Linear(256, n_classes)
            )
        
        def forward(self, x, glcm_feats):
            x = self.base_model(x)
            combined = torch.cat((x, glcm_feats), dim=1)
            return self.classifier(combined)
    
    # Initialize model
    model = TrialCombinedClassifier(n_classes=n_classes).to(device)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.SGD(model.parameters(), lr=lr, momentum=momentum)
    
    # Training loop (short for optimization)
    model.train()
    for epoch in range(3):  # Few epochs for quick evaluation
        for batch_idx, (inputs, glcm_feats, labels) in enumerate(train_loader):
            if batch_idx > 10:  # Limit batches for speed
                break
                
            inputs = inputs.to(device, non_blocking=True)
            glcm_feats = glcm_feats.to(device, non_blocking=True)
            labels = labels.to(device, non_blocking=True)
            
            optimizer.zero_grad()
            outputs = model(inputs, glcm_feats)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
    
    # Validation evaluation
    model.eval()
    correct = 0
    total = 0
    
    with torch.no_grad():
        for batch_idx, (inputs, glcm_feats, labels) in enumerate(val_loader):
            if batch_idx > 15:  # Limit for speed
                break
                
            inputs = inputs.to(device, non_blocking=True)
            glcm_feats = glcm_feats.to(device, non_blocking=True)
            labels = labels.to(device, non_blocking=True)
            
            outputs = model(inputs, glcm_feats)
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
    
    accuracy = correct / total if total > 0 else 0
    print(f"   ğŸ¯ Trial {trial.number} accuracy: {accuracy:.4f}")
    
    # Clean up memory
    del model
    torch.cuda.empty_cache()
    
    return 1 - accuracy  # Optuna minimizes, so return 1 - accuracy

def run_optuna_optimization(train_loader, val_loader, n_classes, n_trials=20):
    """Run Optuna hyperparameter optimization"""
    
    print("ğŸ”§ Starting Optuna hyperparameter optimization...")
    print(f"   ğŸ¯ Target: {n_trials} trials")
    print(f"   ğŸ“Š Classes: {n_classes}")
    
    # Create Optuna study
    study = optuna.create_study(
        direction='minimize',
        study_name='densenet_glcm_optimization',
        pruner=optuna.pruners.MedianPruner(n_startup_trials=5, n_warmup_steps=2)
    )
    
    # Run optimization
    start_time = time.time()
    study.optimize(
        lambda trial: objective_densenet_glcm(trial, train_loader, val_loader, n_classes),
        n_trials=n_trials,
        timeout=1800  # 30 minutes max
    )
    optimization_time = time.time() - start_time
    
    print(f"âœ… Optimization completed in {optimization_time:.2f}s ({optimization_time/60:.1f} minutes)")
    
    # Best parameters
    best_params = study.best_params
    best_score = 1 - study.best_value  # Convert back to accuracy
    
    print(f"\nğŸ† BEST HYPERPARAMETERS:")
    print(f"   ğŸ¯ Best Validation Accuracy: {best_score:.4f}")
    for param, value in best_params.items():
        print(f"   ğŸ”§ {param}: {value}")
    
    return best_params, best_score

def train_final_model(train_loader, val_loader, test_loader, class_names, best_params, num_epochs=50):
    """Train final model with optimized hyperparameters"""
    
    print("ğŸ‹ï¸  Training final model with optimized hyperparameters...")
    print(f"ğŸ“Š Using parameters: {best_params}")
    
    # Create model with best parameters
    class OptimizedCombinedClassifier(nn.Module):
        def __init__(self, n_classes, glcm_size=12):
            super(OptimizedCombinedClassifier, self).__init__()
            self.base_model = models.densenet121(pretrained=True)
            self.base_model.classifier = nn.Identity()
            self.feature_size = 1024
            self.glcm_size = glcm_size
            
            self.classifier = nn.Sequential(
                nn.BatchNorm1d(self.feature_size + self.glcm_size, momentum=0.999, eps=0.001),
                nn.Linear(self.feature_size + self.glcm_size, 1024),
                nn.ReLU(),
                nn.Dropout(best_params.get('dropout_1', 0.2)),
                nn.Linear(1024, 512),
                nn.ReLU(),
                nn.Dropout(best_params.get('dropout_2', 0.3)),
                nn.Linear(512, 256),
                nn.ReLU(),
                nn.Dropout(best_params.get('dropout_3', 0.4)),
                nn.Linear(256, n_classes)
            )
        
        def forward(self, x, glcm_feats):
            x = self.base_model(x)
            combined = torch.cat((x, glcm_feats), dim=1)
            return self.classifier(combined)
    
    # Initialize model and training components
    model = OptimizedCombinedClassifier(n_classes=len(class_names)).to(device)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.SGD(
        model.parameters(),
        lr=best_params.get('lr', 0.001),
        momentum=best_params.get('momentum', 0.9)
    )
    
    # Learning rate scheduler
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='max', patience=5, factor=0.5)
    
    # Training tracking
    train_accs, val_accs, train_losses, val_losses = [], [], [], []
    best_val_acc = 0.0
    patience_counter = 0
    patience = 10
    
    print("ğŸš€ Starting training...")
    print("=" * 80)
    
    for epoch in range(num_epochs):
        epoch_start = time.time()
        
        # Training phase
        model.train()
        running_loss = 0.0
        correct, total = 0, 0
        
        train_pbar = tqdm(train_loader, desc=f"ğŸƒ Epoch {epoch+1}/{num_epochs}", leave=False)
        for inputs, glcm_feats, labels in train_pbar:
            inputs = inputs.to(device, non_blocking=True)
            glcm_feats = glcm_feats.to(device, non_blocking=True)
            labels = labels.to(device, non_blocking=True)
            
            optimizer.zero_grad()
            outputs = model(inputs, glcm_feats)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            
            running_loss += loss.item()
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
            
            train_pbar.set_postfix({
                'Loss': f'{loss.item():.4f}',
                'Acc': f'{correct/total:.4f}'
            })
        
        train_acc = correct / total
        train_loss = running_loss / len(train_loader)
        train_accs.append(train_acc)
        train_losses.append(train_loss)
        
        # Validation phase
        model.eval()
        val_correct, val_total = 0, 0
        val_loss = 0.0
        
        with torch.no_grad():
            for inputs, glcm_feats, labels in val_loader:
                inputs = inputs.to(device, non_blocking=True)
                glcm_feats = glcm_feats.to(device, non_blocking=True)
                labels = labels.to(device, non_blocking=True)
                
                outputs = model(inputs, glcm_feats)
                loss = criterion(outputs, labels)
                
                val_loss += loss.item()
                _, predicted = torch.max(outputs, 1)
                val_total += labels.size(0)
                val_correct += (predicted == labels).sum().item()
        
        val_acc = val_correct / val_total
        val_loss = val_loss / len(val_loader)
        val_accs.append(val_acc)
        val_losses.append(val_loss)
        
        # Learning rate scheduling
        scheduler.step(val_acc)
        
        epoch_time = time.time() - epoch_start
        
        print(f"â±ï¸  Epoch {epoch+1:2d} ({epoch_time:5.1f}s) | "
              f"Train: {train_acc:.4f} | Val: {val_acc:.4f} | "
              f"LR: {optimizer.param_groups[0]['lr']:.6f}")
        
        # Save best model
        if val_acc > best_val_acc:
            best_val_acc = val_acc
            patience_counter = 0
            torch.save(model.state_dict(), 'best_densenet_glcm_model.pth')
            print(f"ğŸ¯ New best validation accuracy: {val_acc:.4f} - Model saved!")
        else:
            patience_counter += 1
            
        if patience_counter >= patience:
            print(f"ğŸ›‘ Early stopping after {epoch+1} epochs")
            break
    
    # Load best model
    model.load_state_dict(torch.load('best_densenet_glcm_model.pth'))
    
    # Final evaluation on test set
    print("\nğŸ§ª Final evaluation on test set...")
    model.eval()
    all_preds = []
    all_labels = []
    
    with torch.no_grad():
        for inputs, glcm_feats, labels in tqdm(test_loader, desc="Testing"):
            inputs = inputs.to(device, non_blocking=True)
            glcm_feats = glcm_feats.to(device, non_blocking=True)
            
            outputs = model(inputs, glcm_feats)
            _, predicted = torch.max(outputs, 1)
            
            all_preds.extend(predicted.cpu().numpy())
            all_labels.extend(labels.numpy())
    
    test_accuracy = accuracy_score(all_labels, all_preds)
    
    print(f"ğŸ¯ Final Test Accuracy: {test_accuracy:.4f}")
    print("\nğŸ“Š Classification Report:")
    print(classification_report(all_labels, all_preds, target_names=class_names))
    
    # Plot results
    plot_training_results(train_accs, val_accs, train_losses, val_losses, test_accuracy, class_names, all_labels, all_preds)
    
    return model, test_accuracy, best_val_acc

def plot_training_results(train_accs, val_accs, train_losses, val_losses, test_accuracy, class_names, all_labels, all_preds):
    """Plot training curves and confusion matrix"""
    
    plt.figure(figsize=(15, 5))
    
    # Loss curves
    plt.subplot(1, 3, 1)
    plt.plot(train_losses, label='Train Loss', color='blue')
    plt.plot(val_losses, label='Val Loss', color='red')
    plt.title('Loss Curves')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    
    # Accuracy curves
    plt.subplot(1, 3, 2)
    plt.plot(train_accs, label='Train Acc', color='blue')
    plt.plot(val_accs, label='Val Acc', color='red')
    plt.title('Accuracy Curves')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True)
    
    # Validation accuracy progress
    plt.subplot(1, 3, 3)
    plt.plot([acc * 100 for acc in val_accs], 'g-', linewidth=2)
    plt.axhline(y=test_accuracy*100, color='r', linestyle='--', label=f'Test Acc: {test_accuracy:.3f}')
    plt.title('Validation Accuracy Progress')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy (%)')
    plt.legend()
    plt.grid(True)
    
    plt.tight_layout()
    plt.show()
    
    # Confusion Matrix
    cm = confusion_matrix(all_labels, all_preds)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=class_names, yticklabels=class_names)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title(f'Confusion Matrix - Test Accuracy: {test_accuracy:.4f}')
    plt.tight_layout()
    plt.show()

# Main execution
if __name__ == "__main__":
    print("ğŸš€ DenseNet121 + GLCM with Optuna Optimization for Folder-based DR Dataset")
    print("=" * 80)
    
    # Configuration
    data_dir = '/kaggle/input/diabetic-retinopathy-224x224-2019-data/colored_images'
    batch_size = 16
    n_trials = 10  # Adjust based on time constraints
    
    print(f"ğŸ“ Dataset directory: {data_dir}")
    print(f"ğŸ”§ Optuna trials: {n_trials}")
    
    # Create data loaders
    print("\nğŸ—ï¸  Creating data loaders...")
    start_time = time.time()
    train_loader, val_loader, test_loader, class_names = create_data_loaders(
        data_dir, batch_size=batch_size, val_split=0.2, test_split=0.1
    )
    loader_time = time.time() - start_time
    print(f"â±ï¸  Data loaders created in {loader_time:.2f}s")
    
    n_classes = len(class_names)
    print(f"ğŸ“Š Dataset summary:")
    print(f"   Classes: {class_names}")
    print(f"   Number of classes: {n_classes}")
    
    # Run Optuna optimization
    print("\nğŸ”§ Running Optuna hyperparameter optimization...")
    best_params, best_score = run_optuna_optimization(
        train_loader, val_loader, n_classes, n_trials=n_trials
    )
    
    # Train final model
    print("\nğŸ‹ï¸  Training final model with optimized hyperparameters...")
    # Train final model
    print("\nğŸ‹ï¸  Training final model with optimized hyperparameters...")
    final_model, test_accuracy, best_val_acc = train_final_model(
        train_loader, val_loader, test_loader, class_names, best_params, num_epochs=50
    )
    
    # Final summary
    print(f"\nğŸ‰ FINAL RESULTS:")
    print("=" * 60)
    print(f"ğŸ”§ Optuna Trials: {n_trials}")
    print(f"ğŸ¯ Best Validation Accuracy: {best_val_acc:.4f} ({best_val_acc*100:.2f}%)")
    print(f"ğŸ§ª Final Test Accuracy: {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")
    print(f"ğŸ“Š Classes: {class_names}")
    print(f"ğŸ—ï¸  Architecture: DenseNet121 + GLCM features")
    print("=" * 60)
    
    print("\nğŸ”§ Optimized Hyperparameters:")
    for param, value in best_params.items():
        print(f"   {param}: {value}")
    
    print("\nâœ… Training completed! Model saved as 'best_densenet_glcm_model.pth'")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset, WeightedRandomSampler
from torchvision import transforms, models
import numpy as np
from PIL import Image
import os
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import time
import optuna
from collections import Counter
import cv2

# GLCM features
from skimage.feature import graycomatrix, graycoprops
from skimage import measure

# Metrics
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

import warnings
warnings.filterwarnings('ignore')

# Set device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"ğŸš€ Using device: {device}")

class GLCMFeatureExtractor:
    """GLCM Feature Extractor optimized for DR classification"""
    
    def __init__(self, distances=[1, 2], angles=[0, 45, 90, 135], levels=32):
        self.distances = distances
        self.angles = [np.deg2rad(angle) for angle in angles]
        self.levels = levels
        print(f"ğŸ”¬ GLCM Extractor: {len(distances)} distances, {len(angles)} angles, {levels} levels")
    
    def extract_glcm_features(self, image):
        """Extract compact GLCM features from PIL image"""
        try:
            # Convert PIL to numpy and preprocess
            if isinstance(image, Image.Image):
                img_array = np.array(image)
            else:
                img_array = image
            
            # Use green channel for better retinal contrast
            if len(img_array.shape) == 3:
                gray = img_array[:, :, 1]  # Green channel
            else:
                gray = img_array
            
            # Resize and enhance
            gray = cv2.resize(gray, (128, 128))  # Smaller for speed
            gray = cv2.equalizeHist(gray)
            
            # Reduce gray levels
            gray = (gray // (256 // self.levels)).astype(np.uint8)
            
            # Extract GLCM features
            features = []
            
            for distance in self.distances:
                glcm = graycomatrix(gray, distances=[distance], angles=self.angles, 
                                 levels=self.levels, symmetric=True, normed=True)
                
                # Core texture properties
                contrast = graycoprops(glcm, 'contrast').flatten()
                homogeneity = graycoprops(glcm, 'homogeneity').flatten()
                energy = graycoprops(glcm, 'energy').flatten()
                
                # Add mean and std for each property
                features.extend([
                    np.mean(contrast), np.std(contrast),
                    np.mean(homogeneity), np.std(homogeneity),
                    np.mean(energy), np.std(energy)
                ])
            
            return np.array(features, dtype=np.float32)
            
        except Exception as e:
            print(f"âŒ GLCM extraction error: {e}")
            return np.zeros(len(self.distances) * 6, dtype=np.float32)  # 6 features per distance

class FolderDRDataset(Dataset):
    """Dataset for folder-based DR classification with GLCM features"""
    
    def __init__(self, root_dir, transform=None, extract_glcm=True):
        self.root_dir = root_dir
        self.transform = transform
        self.extract_glcm = extract_glcm
        
        if self.extract_glcm:
            self.glcm_extractor = GLCMFeatureExtractor()
        
        # Load samples from folders
        self.samples = []
        self.classes = []
        self.class_to_idx = {}
        
        print(f"ğŸ“ Loading dataset from: {root_dir}")
        
        # Get class folders
        for class_name in sorted(os.listdir(root_dir)):
            class_path = os.path.join(root_dir, class_name)
            if os.path.isdir(class_path):
                self.classes.append(class_name)
                class_idx = len(self.classes) - 1
                self.class_to_idx[class_name] = class_idx
                
                # Load images from this class folder
                image_count = 0
                for img_name in os.listdir(class_path):
                    if img_name.lower().endswith(('.png', '.jpg', '.jpeg')):
                        img_path = os.path.join(class_path, img_name)
                        self.samples.append((img_path, class_idx))
                        image_count += 1
                
                print(f"   ğŸ“Š Class {class_name} (ID: {class_idx}): {image_count} images")
        
        print(f"âœ… Loaded {len(self.samples)} total samples from {len(self.classes)} classes")
        
        # Calculate class distribution
        labels = [label for _, label in self.samples]
        self.class_counts = Counter(labels)
        
        # Calculate sample weights for balanced sampling
        total_samples = len(self.samples)
        self.class_weights = {cls_idx: total_samples / count 
                             for cls_idx, count in self.class_counts.items()}
        self.sample_weights = [self.class_weights[label] for _, label in self.samples]
    
    def __len__(self):
        return len(self.samples)
    
    def __getitem__(self, idx):
        img_path, label = self.samples[idx]
        
        # Load image
        try:
            image = Image.open(img_path).convert('RGB')
        except Exception as e:
            print(f"âŒ Error loading {img_path}: {e}")
            image = Image.new('RGB', (224, 224), (0, 0, 0))
        
        # Extract GLCM features before transform
        if self.extract_glcm:
            glcm_features = self.glcm_extractor.extract_glcm_features(image)
        else:
            glcm_features = torch.zeros(12, dtype=torch.float32)  # Default size
        
        # Apply transforms
        if self.transform:
            image = self.transform(image)
        
        return image, torch.tensor(glcm_features, dtype=torch.float32), label

class CombinedResNet50Classifier(nn.Module):
    """ResNet50 + GLCM Combined Classifier"""
    
    def __init__(self, n_classes, glcm_size=12):
        super(CombinedResNet50Classifier, self).__init__()
        print(f"ğŸ—ï¸  Building ResNet50 + GLCM model...")
        
        # ResNet50 backbone
        self.base_model = models.resnet50(pretrained=True)
        self.base_model.fc = nn.Identity()  # Remove final FC layer
        self.feature_size = 2048  # ResNet50 feature size
        self.glcm_size = glcm_size
        
        # Combined classifier
        self.classifier = nn.Sequential(
            nn.BatchNorm1d(self.feature_size + self.glcm_size, momentum=0.999, eps=0.001),
            nn.Linear(self.feature_size + self.glcm_size, 1024),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(1024, 512),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(512, 256),
            nn.ReLU(),
            nn.Dropout(0.4),
            nn.Linear(256, n_classes)
        )
        
        print(f"âœ… ResNet50 Model built: {self.feature_size} CNN + {self.glcm_size} GLCM â†’ {n_classes} classes")
    
    def forward(self, x, glcm_feats):
        # ResNet50 features
        x = self.base_model(x)
        
        # Combine CNN and GLCM features
        combined = torch.cat((x, glcm_feats), dim=1)
        
        # Final classification
        return self.classifier(combined)

def create_data_loaders(data_dir, batch_size=16, val_split=0.2, test_split=0.1):
    """Create train/val/test data loaders from folder structure"""
    
    print("ğŸ¨ Setting up data transformations...")
    
    # Data transformations
    train_transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.RandomHorizontalFlip(p=0.5),
        transforms.RandomRotation(degrees=10),
        transforms.ColorJitter(brightness=0.2, contrast=0.2),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])
    
    val_test_transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])
    
    # Create full dataset
    print("ğŸ“¦ Creating full dataset...")
    full_dataset = FolderDRDataset(data_dir, transform=None, extract_glcm=True)
    
    # Create stratified splits
    print("âœ‚ï¸  Creating stratified train/val/test splits...")
    
    # Group samples by class
    class_samples = {i: [] for i in range(len(full_dataset.classes))}
    for idx, (_, label) in enumerate(full_dataset.samples):
        class_samples[label].append(idx)
    
    train_indices, val_indices, test_indices = [], [], []
    
    for class_id, samples in class_samples.items():
        np.random.shuffle(samples)
        n_samples = len(samples)
        
        # Calculate split sizes
        test_size = max(1, int(test_split * n_samples))
        val_size = max(1, int(val_split * n_samples))
        train_size = n_samples - test_size - val_size
        
        # Split indices
        test_indices.extend(samples[:test_size])
        val_indices.extend(samples[test_size:test_size + val_size])
        train_indices.extend(samples[test_size + val_size:])
    
    # Shuffle indices
    np.random.shuffle(train_indices)
    np.random.shuffle(val_indices)
    np.random.shuffle(test_indices)
    
    print(f"ğŸ“Š Data split:")
    print(f"   Training:   {len(train_indices)} samples")
    print(f"   Validation: {len(val_indices)} samples") 
    print(f"   Test:       {len(test_indices)} samples")
    
    # Create subset datasets with transforms
    class TransformSubset(Dataset):
        def __init__(self, dataset, indices, transform):
            self.dataset = dataset
            self.indices = indices
            self.transform = transform
        
        def __len__(self):
            return len(self.indices)
        
        def __getitem__(self, idx):
            original_idx = self.indices[idx]
            img_path, label = self.dataset.samples[original_idx]
            
            # Load image
            try:
                image = Image.open(img_path).convert('RGB')
            except:
                image = Image.new('RGB', (224, 224), (0, 0, 0))
            
            # Extract GLCM features
            glcm_features = self.dataset.glcm_extractor.extract_glcm_features(image)
            
            # Apply transform
            if self.transform:
                image = self.transform(image)
            
            return image, torch.tensor(glcm_features, dtype=torch.float32), label
    
    # Create transformed datasets
    train_dataset = TransformSubset(full_dataset, train_indices, train_transform)
    val_dataset = TransformSubset(full_dataset, val_indices, val_test_transform)
    test_dataset = TransformSubset(full_dataset, test_indices, val_test_transform)
    
    # Create weighted sampler for training
    train_weights = [full_dataset.sample_weights[i] for i in train_indices]
    train_sampler = WeightedRandomSampler(train_weights, len(train_weights), replacement=True)
    
    # Create data loaders
    print("ğŸšš Creating data loaders...")
    train_loader = DataLoader(train_dataset, batch_size=batch_size, sampler=train_sampler,
                             num_workers=2, pin_memory=True)
    val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False,
                           num_workers=2, pin_memory=True)
    test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False,
                            num_workers=2, pin_memory=True)
    
    return train_loader, val_loader, test_loader, full_dataset.classes

def objective_resnet50_glcm(trial, train_loader, val_loader, n_classes):
    """Optuna objective function for ResNet50 + GLCM hyperparameter optimization"""
    
    # Suggest hyperparameters
    lr = trial.suggest_loguniform("lr", 1e-5, 1e-2)
    momentum = trial.suggest_uniform("momentum", 0.5, 0.99)
    dropout_1 = trial.suggest_uniform("dropout_1", 0.1, 0.4)
    dropout_2 = trial.suggest_uniform("dropout_2", 0.2, 0.5)
    dropout_3 = trial.suggest_uniform("dropout_3", 0.3, 0.6)
    
    print(f"ğŸ”§ Trial {trial.number}: lr={lr:.6f}, momentum={momentum:.3f}")
    
    # Create model with trial-specific dropout
    class TrialResNet50Classifier(nn.Module):
        def __init__(self, n_classes, glcm_size=12):
            super(TrialResNet50Classifier, self).__init__()
            self.base_model = models.resnet50(pretrained=True)
            self.base_model.fc = nn.Identity()
            self.feature_size = 2048
            self.glcm_size = glcm_size
            
            self.classifier = nn.Sequential(
                nn.BatchNorm1d(self.feature_size + self.glcm_size, momentum=0.999, eps=0.001),
                nn.Linear(self.feature_size + self.glcm_size, 1024),
                nn.ReLU(),
                nn.Dropout(dropout_1),
                nn.Linear(1024, 512),
                nn.ReLU(),
                nn.Dropout(dropout_2),
                nn.Linear(512, 256),
                nn.ReLU(),
                nn.Dropout(dropout_3),
                nn.Linear(256, n_classes)
            )
        
        def forward(self, x, glcm_feats):
            x = self.base_model(x)
            combined = torch.cat((x, glcm_feats), dim=1)
            return self.classifier(combined)
    
    # Initialize model
    model = TrialResNet50Classifier(n_classes=n_classes).to(device)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.SGD(model.parameters(), lr=lr, momentum=momentum)
    
    # Training loop (short for optimization)
    model.train()
    for epoch in range(3):  # Few epochs for quick evaluation
        for batch_idx, (inputs, glcm_feats, labels) in enumerate(train_loader):
            if batch_idx > 8:  # Limit batches for speed
                break
                
            inputs = inputs.to(device, non_blocking=True)
            glcm_feats = glcm_feats.to(device, non_blocking=True)
            labels = labels.to(device, non_blocking=True)
            
            optimizer.zero_grad()
            outputs = model(inputs, glcm_feats)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
    
    # Validation evaluation
    model.eval()
    correct = 0
    total = 0
    
    with torch.no_grad():
        for batch_idx, (inputs, glcm_feats, labels) in enumerate(val_loader):
            if batch_idx > 12:  # Limit for speed
                break
                
            inputs = inputs.to(device, non_blocking=True)
            glcm_feats = glcm_feats.to(device, non_blocking=True)
            labels = labels.to(device, non_blocking=True)
            
            outputs = model(inputs, glcm_feats)
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
    
    accuracy = correct / total if total > 0 else 0
    print(f"   ğŸ¯ Trial {trial.number} accuracy: {accuracy:.4f}")
    
    # Clean up memory
    del model
    torch.cuda.empty_cache()
    
    return 1 - accuracy  # Optuna minimizes, so return 1 - accuracy

def run_optuna_optimization(train_loader, val_loader, n_classes, n_trials=15):
    """Run Optuna hyperparameter optimization for ResNet50"""
    
    print("ğŸ”§ Starting Optuna hyperparameter optimization for ResNet50...")
    print(f"   ğŸ¯ Target: {n_trials} trials")
    print(f"   ğŸ“Š Classes: {n_classes}")
    
    # Create Optuna study
    study = optuna.create_study(
        direction='minimize',
        study_name='resnet50_glcm_optimization',
        pruner=optuna.pruners.MedianPruner(n_startup_trials=3, n_warmup_steps=1)
    )
    
    # Run optimization
    start_time = time.time()
    study.optimize(
        lambda trial: objective_resnet50_glcm(trial, train_loader, val_loader, n_classes),
        n_trials=n_trials,
        timeout=1200  # 20 minutes max
    )
    optimization_time = time.time() - start_time
    
    print(f"âœ… Optimization completed in {optimization_time:.2f}s ({optimization_time/60:.1f} minutes)")
    
    # Best parameters
    best_params = study.best_params
    best_score = 1 - study.best_value  # Convert back to accuracy
    
    print(f"\nğŸ† BEST HYPERPARAMETERS FOR RESNET50:")
    print(f"   ğŸ¯ Best Validation Accuracy: {best_score:.4f}")
    for param, value in best_params.items():
        print(f"   ğŸ”§ {param}: {value}")
    
    return best_params, best_score

def train_final_resnet50_model(train_loader, val_loader, test_loader, class_names, best_params, num_epochs=50):
    """Train final ResNet50 model with optimized hyperparameters"""
    
    print("ğŸ‹ï¸  Training final ResNet50 model with optimized hyperparameters...")
    print(f"ğŸ“Š Using parameters: {best_params}")
    
    # Create model with best parameters
    class OptimizedResNet50Classifier(nn.Module):
        def __init__(self, n_classes, glcm_size=12):
            super(OptimizedResNet50Classifier, self).__init__()
            self.base_model = models.resnet50(pretrained=True)
            self.base_model.fc = nn.Identity()
            self.feature_size = 2048
            self.glcm_size = glcm_size
            
            self.classifier = nn.Sequential(
                nn.BatchNorm1d(self.feature_size + self.glcm_size, momentum=0.999, eps=0.001),
                nn.Linear(self.feature_size + self.glcm_size, 1024),
                nn.ReLU(),
                nn.Dropout(best_params.get('dropout_1', 0.2)),
                nn.Linear(1024, 512),
                nn.ReLU(),
                nn.Dropout(best_params.get('dropout_2', 0.3)),
                nn.Linear(512, 256),
                nn.ReLU(),
                nn.Dropout(best_params.get('dropout_3', 0.4)),
                nn.Linear(256, n_classes)
            )
        
        def forward(self, x, glcm_feats):
            x = self.base_model(x)
            combined = torch.cat((x, glcm_feats), dim=1)
            return self.classifier(combined)
    
    # Initialize model and training components
    model = OptimizedResNet50Classifier(n_classes=len(class_names)).to(device)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.SGD(
        model.parameters(),
        lr=best_params.get('lr', 0.001),
        momentum=best_params.get('momentum', 0.9)
    )
    
    # Learning rate scheduler
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='max', patience=5, factor=0.5)
    
    # Training tracking
    train_accs, val_accs, train_losses, val_losses = [], [], [], []
    best_val_acc = 0.0
    patience_counter = 0
    patience = 10
    
    print("ğŸš€ Starting ResNet50 training...")
    print("=" * 80)
    
    for epoch in range(num_epochs):
        epoch_start = time.time()
        
        # Training phase
        model.train()
        running_loss = 0.0
        correct, total = 0, 0
        
        train_pbar = tqdm(train_loader, desc=f"ğŸƒ Epoch {epoch+1}/{num_epochs}", leave=False)
        for inputs, glcm_feats, labels in train_pbar:
            inputs = inputs.to(device, non_blocking=True)
            glcm_feats = glcm_feats.to(device, non_blocking=True)
            labels = labels.to(device, non_blocking=True)
            
            optimizer.zero_grad()
            outputs = model(inputs, glcm_feats)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            
            running_loss += loss.item()
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
            
            train_pbar.set_postfix({
                'Loss': f'{loss.item():.4f}',
                'Acc': f'{correct/total:.4f}'
            })
        
        train_acc = correct / total
        train_loss = running_loss / len(train_loader)
        train_accs.append(train_acc)
        train_losses.append(train_loss)
        
        # Validation phase
        model.eval()
        val_correct, val_total = 0, 0
        val_loss = 0.0
        
        with torch.no_grad():
            for inputs, glcm_feats, labels in val_loader:
                inputs = inputs.to(device, non_blocking=True)
                glcm_feats = glcm_feats.to(device, non_blocking=True)
                labels = labels.to(device, non_blocking=True)
                
                outputs = model(inputs, glcm_feats)
                loss = criterion(outputs, labels)
                
                val_loss += loss.item()
                _, predicted = torch.max(outputs, 1)
                val_total += labels.size(0)
                val_correct += (predicted == labels).sum().item()
        
        val_acc = val_correct / val_total
        val_loss = val_loss / len(val_loader)
        val_accs.append(val_acc)
        val_losses.append(val_loss)
        
        # Learning rate scheduling
        scheduler.step(val_acc)
        
        epoch_time = time.time() - epoch_start
        
        print(f"â±ï¸  Epoch {epoch+1:2d} ({epoch_time:5.1f}s) | "
              f"Train: {train_acc:.4f} | Val: {val_acc:.4f} | "
              f"LR: {optimizer.param_groups[0]['lr']:.6f}")
        
        # Save best model
        if val_acc > best_val_acc:
            best_val_acc = val_acc
            patience_counter = 0
            torch.save(model.state_dict(), 'best_resnet50_glcm_model.pth')
            print(f"ğŸ¯ New best validation accuracy: {val_acc:.4f} - Model saved!")
        else:
            patience_counter += 1
            
        if patience_counter >= patience:
            print(f"ğŸ›‘ Early stopping after {epoch+1} epochs")
            break
    
    # Load best model
    model.load_state_dict(torch.load('best_resnet50_glcm_model.pth'))
    
    # Final evaluation on test set
    print("\nğŸ§ª Final evaluation on test set...")
    model.eval()
    all_preds = []
    all_labels = []
    
    with torch.no_grad():
        for inputs, glcm_feats, labels in tqdm(test_loader, desc="Testing"):
            inputs = inputs.to(device, non_blocking=True)
            glcm_feats = glcm_feats.to(device, non_blocking=True)
            
            outputs = model(inputs, glcm_feats)
            _, predicted = torch.max(outputs, 1)
            
            all_preds.extend(predicted.cpu().numpy())
            all_labels.extend(labels.numpy())
    
    test_accuracy = accuracy_score(all_labels, all_preds)
    
    print(f"ğŸ¯ Final Test Accuracy: {test_accuracy:.4f}")
    print("\nğŸ“Š Classification Report:")
    print(classification_report(all_labels, all_preds, target_names=class_names))
    
    # Plot results
    plot_training_results(train_accs, val_accs, train_losses, val_losses, test_accuracy, class_names, all_labels, all_preds)
    
    return model, test_accuracy, best_val_acc

def plot_training_results(train_accs, val_accs, train_losses, val_losses, test_accuracy, class_names, all_labels, all_preds):
    """Plot training curves and confusion matrix"""
    
    plt.figure(figsize=(15, 5))
    
    # Loss curves
    plt.subplot(1, 3, 1)
    plt.plot(train_losses, label='Train Loss', color='blue')
    plt.plot(val_losses, label='Val Loss', color='red')
    plt.title('ResNet50 Loss Curves')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    
    # Accuracy curves
    plt.subplot(1, 3, 2)
    plt.plot(train_accs, label='Train Acc', color='blue')
    plt.plot(val_accs, label='Val Acc', color='red')
    plt.title('ResNet50 Accuracy Curves')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True)
    
    # Validation accuracy progress
    plt.subplot(1, 3, 3)
    plt.plot([acc * 100 for acc in val_accs], 'g-', linewidth=2)
    plt.axhline(y=test_accuracy*100, color='r', linestyle='--', label=f'Test Acc: {test_accuracy:.3f}')
    plt.title('ResNet50 Validation Accuracy Progress')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy (%)')
    plt.legend()
    plt.grid(True)
    
    plt.tight_layout()
    plt.show()
    
    # Confusion Matrix
    cm = confusion_matrix(all_labels, all_preds)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=class_names, yticklabels=class_names)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title(f'ResNet50 + GLCM Confusion Matrix - Test Accuracy: {test_accuracy:.4f}')
    plt.tight_layout()
    plt.show()

# Main execution
if __name__ == "__main__":
    print("ğŸš€ ResNet50 + GLCM with Optuna Optimization for Folder-based DR Dataset")
    print("=" * 80)
    
    # Configuration
    data_dir = '/kaggle/input/diabetic-retinopathy-224x224-2019-data/colored_images'
    batch_size = 16
    n_trials = 10  # Adjust based on time constraints
    
    print(f"ğŸ“ Dataset directory: {data_dir}")
    print(f"ğŸ”§ Optuna trials: {n_trials}")
    
    # Create data loaders
    print("\nğŸ—ï¸  Creating data loaders...")
    start_time = time.time()
    train_loader, val_loader, test_loader, class_names = create_data_loaders(
        data_dir, batch_size=batch_size, val_split=0.2, test_split=0.1
    )
    loader_time = time.time() - start_time
    print(f"â±ï¸  Data loaders created in {loader_time:.2f}s")
    
    n_classes = len(class_names)
    print(f"ğŸ“Š Dataset summary:")
    print(f"   Classes: {class_names}")
    print(f"   Number of classes: {n_classes}")
    
    # Run Optuna optimization
    print("\nğŸ”§ Running Optuna hyperparameter optimization for ResNet50...")
    best_params, best_score = run_optuna_optimization(
        train_loader, val_loader, n_classes, n_trials=n_trials
    )
    
    # Train final model
    print("\nğŸ‹ï¸  Training final ResNet50 model with optimized hyperparameters...")
    print("\nğŸ‹ï¸  Training final model with optimized hyperparameters...")
    final_model, test_accuracy, best_val_acc = train_final_model(
        train_loader, val_loader, test_loader, class_names, best_params, num_epochs=50
    )
    
    # Final summary
    print(f"\nğŸ‰ FINAL RESULTS:")
    print("=" * 60)
    print(f"ğŸ”§ Optuna Trials: {n_trials}")
    print(f"ğŸ¯ Best Validation Accuracy: {best_val_acc:.4f} ({best_val_acc*100:.2f}%)")
    print(f"ğŸ§ª Final Test Accuracy: {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")
    print(f"ğŸ“Š Classes: {class_names}")
    print(f"ğŸ—ï¸  Architecture: Resnet50 + GLCM features")
    print("=" * 60)
    
    print("\nğŸ”§ Optimized Hyperparameters:")
    for param, value in best_params.items():
        print(f"   {param}: {value}")
    
    print("\nâœ… Training completed! Model saved as 'best_resnet_glcm_model.pth'")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset, WeightedRandomSampler
from torchvision import transforms, models
import numpy as np
from PIL import Image
import os
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import time
import optuna
from collections import Counter
import cv2

# GLCM features
from skimage.feature import graycomatrix, graycoprops
from skimage import measure

# Metrics
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Try to import EfficientNet
try:
    from efficientnet_pytorch import EfficientNet
    EFFICIENTNET_AVAILABLE = True
    print("âœ… Using efficientnet_pytorch library")
except ImportError:
    EFFICIENTNET_AVAILABLE = False
    print("âš ï¸  efficientnet_pytorch not available, using torchvision version")

import warnings
warnings.filterwarnings('ignore')

# Set device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"ğŸš€ Using device: {device}")

class GLCMFeatureExtractor:
    """GLCM Feature Extractor optimized for DR classification"""
    
    def __init__(self, distances=[1, 2], angles=[0, 45, 90, 135], levels=32):
        self.distances = distances
        self.angles = [np.deg2rad(angle) for angle in angles]
        self.levels = levels
        print(f"ğŸ”¬ GLCM Extractor: {len(distances)} distances, {len(angles)} angles, {levels} levels")
    
    def extract_glcm_features(self, image):
        """Extract compact GLCM features from PIL image"""
        try:
            # Convert PIL to numpy and preprocess
            if isinstance(image, Image.Image):
                img_array = np.array(image)
            else:
                img_array = image
            
            # Use green channel for better retinal contrast
            if len(img_array.shape) == 3:
                gray = img_array[:, :, 1]  # Green channel
            else:
                gray = img_array
            
            # Resize and enhance
            gray = cv2.resize(gray, (128, 128))  # Smaller for speed
            gray = cv2.equalizeHist(gray)
            
            # Reduce gray levels
            gray = (gray // (256 // self.levels)).astype(np.uint8)
            
            # Extract GLCM features
            features = []
            
            for distance in self.distances:
                glcm = graycomatrix(gray, distances=[distance], angles=self.angles, 
                                 levels=self.levels, symmetric=True, normed=True)
                
                # Core texture properties
                contrast = graycoprops(glcm, 'contrast').flatten()
                homogeneity = graycoprops(glcm, 'homogeneity').flatten()
                energy = graycoprops(glcm, 'energy').flatten()
                
                # Add mean and std for each property
                features.extend([
                    np.mean(contrast), np.std(contrast),
                    np.mean(homogeneity), np.std(homogeneity),
                    np.mean(energy), np.std(energy)
                ])
            
            return np.array(features, dtype=np.float32)
            
        except Exception as e:
            print(f"âŒ GLCM extraction error: {e}")
            return np.zeros(len(self.distances) * 6, dtype=np.float32)  # 6 features per distance

class FolderDRDataset(Dataset):
    """Dataset for folder-based DR classification with GLCM features"""
    
    def __init__(self, root_dir, transform=None, extract_glcm=True):
        self.root_dir = root_dir
        self.transform = transform
        self.extract_glcm = extract_glcm
        
        if self.extract_glcm:
            self.glcm_extractor = GLCMFeatureExtractor()
        
        # Load samples from folders
        self.samples = []
        self.classes = []
        self.class_to_idx = {}
        
        print(f"ğŸ“ Loading dataset from: {root_dir}")
        
        # Get class folders
        for class_name in sorted(os.listdir(root_dir)):
            class_path = os.path.join(root_dir, class_name)
            if os.path.isdir(class_path):
                self.classes.append(class_name)
                class_idx = len(self.classes) - 1
                self.class_to_idx[class_name] = class_idx
                
                # Load images from this class folder
                image_count = 0
                for img_name in os.listdir(class_path):
                    if img_name.lower().endswith(('.png', '.jpg', '.jpeg')):
                        img_path = os.path.join(class_path, img_name)
                        self.samples.append((img_path, class_idx))
                        image_count += 1
                
                print(f"   ğŸ“Š Class {class_name} (ID: {class_idx}): {image_count} images")
        
        print(f"âœ… Loaded {len(self.samples)} total samples from {len(self.classes)} classes")
        
        # Calculate class distribution
        labels = [label for _, label in self.samples]
        self.class_counts = Counter(labels)
        
        # Calculate sample weights for balanced sampling
        total_samples = len(self.samples)
        self.class_weights = {cls_idx: total_samples / count 
                             for cls_idx, count in self.class_counts.items()}
        self.sample_weights = [self.class_weights[label] for _, label in self.samples]
    
    def __len__(self):
        return len(self.samples)
    
    def __getitem__(self, idx):
        img_path, label = self.samples[idx]
        
        # Load image
        try:
            image = Image.open(img_path).convert('RGB')
        except Exception as e:
            print(f"âŒ Error loading {img_path}: {e}")
            image = Image.new('RGB', (224, 224), (0, 0, 0))
        
        # Extract GLCM features before transform
        if self.extract_glcm:
            glcm_features = self.glcm_extractor.extract_glcm_features(image)
        else:
            glcm_features = torch.zeros(12, dtype=torch.float32)  # Default size
        
        # Apply transforms
        if self.transform:
            image = self.transform(image)
        
        return image, torch.tensor(glcm_features, dtype=torch.float32), label

class CombinedEfficientNetB3Classifier(nn.Module):
    """EfficientNet-B3 + GLCM Combined Classifier"""
    
    def __init__(self, n_classes, glcm_size=12):
        super(CombinedEfficientNetB3Classifier, self).__init__()
        print(f"ğŸ—ï¸  Building EfficientNet-B3 + GLCM model...")
        
        if EFFICIENTNET_AVAILABLE:
            # Use efficientnet_pytorch library
            self.base_model = EfficientNet.from_pretrained('efficientnet-b3')
            self.base_model._fc = nn.Identity()  # Remove final FC layer
            self.feature_size = 1536  # EfficientNet-B3 feature size
            self.use_pytorch_efficientnet = True
            print("âœ… Using efficientnet_pytorch library")
        else:
            # Use torchvision version
            try:
                self.base_model = models.efficientnet_b3(pretrained=True)
                self.base_model.classifier = nn.Identity()
                self.feature_size = 1536
                self.use_pytorch_efficientnet = False
                print("âœ… Using torchvision EfficientNet-B3")
            except:
                # Fallback to ResNet34 if EfficientNet not available
                print("âš ï¸  EfficientNet-B3 not available, using ResNet34 as fallback...")
                self.base_model = models.resnet34(pretrained=True)
                self.base_model.fc = nn.Identity()
                self.feature_size = 512
                self.use_pytorch_efficientnet = False
        
        self.glcm_size = glcm_size
        
        # Combined classifier
        self.classifier = nn.Sequential(
            nn.BatchNorm1d(self.feature_size + self.glcm_size, momentum=0.999, eps=0.001),
            nn.Linear(self.feature_size + self.glcm_size, 1024),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(1024, 512),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(512, 256),
            nn.ReLU(),
            nn.Dropout(0.4),
            nn.Linear(256, n_classes)
        )
        
        print(f"âœ… EfficientNet-B3 Model built: {self.feature_size} CNN + {self.glcm_size} GLCM â†’ {n_classes} classes")
    
    def forward(self, x, glcm_feats):
        # EfficientNet features
        if EFFICIENTNET_AVAILABLE and self.use_pytorch_efficientnet:
            # For efficientnet_pytorch library
            x = self.base_model.extract_features(x)
            x = self.base_model._avg_pooling(x)
            x = x.view(x.size(0), -1)
        else:
            # For torchvision version or fallback
            x = self.base_model(x)
        
        # Combine CNN and GLCM features
        combined = torch.cat((x, glcm_feats), dim=1)
        
        # Final classification
        return self.classifier(combined)

def create_efficientnet_transforms():
    """Create EfficientNet-specific transforms with proper image size"""
    
    # EfficientNet-B3 uses 300x300 input size for optimal performance
    img_size = 300
    
    train_transform = transforms.Compose([
        transforms.Resize((img_size, img_size)),
        transforms.RandomHorizontalFlip(p=0.5),
        transforms.RandomRotation(degrees=15),
        transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.1),
        transforms.RandomAffine(degrees=0, translate=(0.1, 0.1)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])
    
    val_test_transform = transforms.Compose([
        transforms.Resize((img_size, img_size)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])
    
    return train_transform, val_test_transform

def create_data_loaders(data_dir, batch_size=12, val_split=0.2, test_split=0.1):
    """Create train/val/test data loaders from folder structure"""
    
    print("ğŸ¨ Setting up EfficientNet-B3 data transformations...")
    
    # EfficientNet-specific transforms
    train_transform, val_test_transform = create_efficientnet_transforms()
    
    # Create full dataset
    print("ğŸ“¦ Creating full dataset...")
    full_dataset = FolderDRDataset(data_dir, transform=None, extract_glcm=True)
    
    # Create stratified splits
    print("âœ‚ï¸  Creating stratified train/val/test splits...")
    
    # Group samples by class
    class_samples = {i: [] for i in range(len(full_dataset.classes))}
    for idx, (_, label) in enumerate(full_dataset.samples):
        class_samples[label].append(idx)
    
    train_indices, val_indices, test_indices = [], [], []
    
    for class_id, samples in class_samples.items():
        np.random.shuffle(samples)
        n_samples = len(samples)
        
        # Calculate split sizes
        test_size = max(1, int(test_split * n_samples))
        val_size = max(1, int(val_split * n_samples))
        train_size = n_samples - test_size - val_size
        
        # Split indices
        test_indices.extend(samples[:test_size])
        val_indices.extend(samples[test_size:test_size + val_size])
        train_indices.extend(samples[test_size + val_size:])
    
    # Shuffle indices
    np.random.shuffle(train_indices)
    np.random.shuffle(val_indices)
    np.random.shuffle(test_indices)
    
    print(f"ğŸ“Š Data split:")
    print(f"   Training:   {len(train_indices)} samples")
    print(f"   Validation: {len(val_indices)} samples") 
    print(f"   Test:       {len(test_indices)} samples")
    
    # Create subset datasets with transforms
    class TransformSubset(Dataset):
        def __init__(self, dataset, indices, transform):
            self.dataset = dataset
            self.indices = indices
            self.transform = transform
        
        def __len__(self):
            return len(self.indices)
        
        def __getitem__(self, idx):
            original_idx = self.indices[idx]
            img_path, label = self.dataset.samples[original_idx]
            
            # Load image
            try:
                image = Image.open(img_path).convert('RGB')
            except:
                image = Image.new('RGB', (300, 300), (0, 0, 0))
            
            # Extract GLCM features
            glcm_features = self.dataset.glcm_extractor.extract_glcm_features(image)
            
            # Apply transform
            if self.transform:
                image = self.transform(image)
            
            return image, torch.tensor(glcm_features, dtype=torch.float32), label
    
    # Create transformed datasets
    train_dataset = TransformSubset(full_dataset, train_indices, train_transform)
    val_dataset = TransformSubset(full_dataset, val_indices, val_test_transform)
    test_dataset = TransformSubset(full_dataset, test_indices, val_test_transform)
    
    # Create weighted sampler for training
    train_weights = [full_dataset.sample_weights[i] for i in train_indices]
    train_sampler = WeightedRandomSampler(train_weights, len(train_weights), replacement=True)
    
    # Create data loaders (smaller batch size for EfficientNet due to larger image size)
    print("ğŸšš Creating data loaders...")
    train_loader = DataLoader(train_dataset, batch_size=batch_size, sampler=train_sampler,
                             num_workers=2, pin_memory=True)
    val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False,
                           num_workers=2, pin_memory=True)
    test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False,
                            num_workers=2, pin_memory=True)
    
    return train_loader, val_loader, test_loader, full_dataset.classes

def objective_efficientnet_b3_glcm(trial, train_loader, val_loader, n_classes):
    """Optuna objective function for EfficientNet-B3 + GLCM hyperparameter optimization"""
    
    # Suggest hyperparameters
    lr = trial.suggest_loguniform("lr", 5e-6, 5e-3)  # Lower LR for EfficientNet
    momentum = trial.suggest_uniform("momentum", 0.6, 0.95)
    dropout_1 = trial.suggest_uniform("dropout_1", 0.1, 0.3)
    dropout_2 = trial.suggest_uniform("dropout_2", 0.2, 0.4)
    dropout_3 = trial.suggest_uniform("dropout_3", 0.3, 0.5)
    weight_decay = trial.suggest_loguniform("weight_decay", 1e-5, 1e-3)
    
    print(f"ğŸ”§ Trial {trial.number}: lr={lr:.6f}, momentum={momentum:.3f}, wd={weight_decay:.6f}")
    
    # Create model with trial-specific dropout
    class TrialEfficientNetB3Classifier(nn.Module):
        def __init__(self, n_classes, glcm_size=12):
            super(TrialEfficientNetB3Classifier, self).__init__()
            
            if EFFICIENTNET_AVAILABLE:
                self.base_model = EfficientNet.from_pretrained('efficientnet-b3')
                self.base_model._fc = nn.Identity()
                self.feature_size = 1536
                self.use_pytorch_efficientnet = True
            else:
                try:
                    self.base_model = models.efficientnet_b3(pretrained=True)
                    self.base_model.classifier = nn.Identity()
                    self.feature_size = 1536
                    self.use_pytorch_efficientnet = False
                except:
                    self.base_model = models.resnet34(pretrained=True)
                    self.base_model.fc = nn.Identity()
                    self.feature_size = 512
                    self.use_pytorch_efficientnet = False
            
            self.glcm_size = glcm_size
            
            self.classifier = nn.Sequential(
                nn.BatchNorm1d(self.feature_size + self.glcm_size, momentum=0.999, eps=0.001),
                nn.Linear(self.feature_size + self.glcm_size, 1024),
                nn.ReLU(),
                nn.Dropout(dropout_1),
                nn.Linear(1024, 512),
                nn.ReLU(),
                nn.Dropout(dropout_2),
                nn.Linear(512, 256),
                nn.ReLU(),
                nn.Dropout(dropout_3),
                nn.Linear(256, n_classes)
            )
        
        def forward(self, x, glcm_feats):
            if EFFICIENTNET_AVAILABLE and self.use_pytorch_efficientnet:
                x = self.base_model.extract_features(x)
                x = self.base_model._avg_pooling(x)
                x = x.view(x.size(0), -1)
            else:
                x = self.base_model(x)
            
            combined = torch.cat((x, glcm_feats), dim=1)
            return self.classifier(combined)
    
    # Initialize model
    model = TrialEfficientNetB3Classifier(n_classes=n_classes).to(device)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.SGD(model.parameters(), lr=lr, momentum=momentum, weight_decay=weight_decay)
    
    # Training loop (short for optimization)
    model.train()
    for epoch in range(3):  # Few epochs for quick evaluation
        for batch_idx, (inputs, glcm_feats, labels) in enumerate(train_loader):
            if batch_idx > 6:  # Limit batches for speed (smaller due to larger images)
                break
                
            inputs = inputs.to(device, non_blocking=True)
            glcm_feats = glcm_feats.to(device, non_blocking=True)
            labels = labels.to(device, non_blocking=True)
            
            optimizer.zero_grad()
            outputs = model(inputs, glcm_feats)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
    
    # Validation evaluation
    model.eval()
    correct = 0
    total = 0
    
    with torch.no_grad():
        for batch_idx, (inputs, glcm_feats, labels) in enumerate(val_loader):
            if batch_idx > 10:  # Limit for speed
                break
                
            inputs = inputs.to(device, non_blocking=True)
            glcm_feats = glcm_feats.to(device, non_blocking=True)
            labels = labels.to(device, non_blocking=True)
            
            outputs = model(inputs, glcm_feats)
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
    
    accuracy = correct / total if total > 0 else 0
    print(f"   ğŸ¯ Trial {trial.number} accuracy: {accuracy:.4f}")
    
    # Clean up memory
    del model
    torch.cuda.empty_cache()
    
    return 1 - accuracy  # Optuna minimizes, so return 1 - accuracy

def run_optuna_optimization(train_loader, val_loader, n_classes, n_trials=12):
    """Run Optuna hyperparameter optimization for EfficientNet-B3"""
    
    print("ğŸ”§ Starting Optuna hyperparameter optimization for EfficientNet-B3...")
    print(f"   ğŸ¯ Target: {n_trials} trials")
    print(f"   ğŸ“Š Classes: {n_classes}")
    
    # Create Optuna study
    study = optuna.create_study(
        direction='minimize',
        study_name='efficientnet_b3_glcm_optimization',
        pruner=optuna.pruners.MedianPruner(n_startup_trials=3, n_warmup_steps=1)
    )
    
    # Run optimization
    start_time = time.time()
    study.optimize(
        lambda trial: objective_efficientnet_b3_glcm(trial, train_loader, val_loader, n_classes),
        n_trials=n_trials,
        timeout=1500  # 25 minutes max
    )
    optimization_time = time.time() - start_time
    
    print(f"âœ… Optimization completed in {optimization_time:.2f}s ({optimization_time/60:.1f} minutes)")
    
    # Best parameters
    best_params = study.best_params
    best_score = 1 - study.best_value  # Convert back to accuracy
    
    print(f"\nğŸ† BEST HYPERPARAMETERS FOR EFFICIENTNET-B3:")
    print(f"   ğŸ¯ Best Validation Accuracy: {best_score:.4f}")
    for param, value in best_params.items():
        print(f"   ğŸ”§ {param}: {value}")
    
    return best_params, best_score

def train_final_efficientnet_b3_model(train_loader, val_loader, test_loader, class_names, best_params, num_epochs=40):
    """Train final EfficientNet-B3 model with optimized hyperparameters"""
    
    print("ğŸ‹ï¸  Training final EfficientNet-B3 model with optimized hyperparameters...")
    print(f"ğŸ“Š Using parameters: {best_params}")
    
    # Create model with best parameters
    class OptimizedEfficientNetB3Classifier(nn.Module):
        def __init__(self, n_classes, glcm_size=12):
            super(OptimizedEfficientNetB3Classifier, self).__init__()
            
            if EFFICIENTNET_AVAILABLE:
                self.base_model = EfficientNet.from_pretrained('efficientnet-b3')
                self.base_model._fc = nn.Identity()
                self.feature_size = 1536
                self.use_pytorch_efficientnet = True
            else:
                try:
                    self.base_model = models.efficientnet_b3(pretrained=True)
                    self.base_model.classifier = nn.Identity()
                    self.feature_size = 1536
                    self.use_pytorch_efficientnet = False
                except:
                    self.base_model = models.resnet34(pretrained=True)
                    self.base_model.fc = nn.Identity()
                    self.feature_size = 512
                    self.use_pytorch_efficientnet = False
            
            self.glcm_size = glcm_size
            
            self.classifier = nn.Sequential(
                nn.BatchNorm1d(self.feature_size + self.glcm_size, momentum=0.999, eps=0.001),
                nn.Linear(self.feature_size + self.glcm_size, 1024),
                nn.ReLU(),
                nn.Dropout(best_params.get('dropout_1', 0.2)),
                nn.Linear(1024, 512),
                nn.ReLU(),
                nn.Dropout(best_params.get('dropout_2', 0.3)),
                nn.Linear(512, 256),
                nn.ReLU(),
                nn.Dropout(best_params.get('dropout_3', 0.4)),
                nn.Linear(256, n_classes)
            )
        
        def forward(self, x, glcm_feats):
            if EFFICIENTNET_AVAILABLE and self.use_pytorch_efficientnet:
                x = self.base_model.extract_features(x)
                x = self.base_model._avg_pooling(x)
                x = x.view(x.size(0), -1)
            else:
                x = self.base_model(x)
            
            combined = torch.cat((x, glcm_feats), dim=1)
            return self.classifier(combined)
    
    # Initialize model and training components
    model = OptimizedEfficientNetB3Classifier(n_classes=len(class_names)).to(device)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.SGD(
        model.parameters(),
        lr=best_params.get('lr', 0.001),
        momentum=best_params.get('momentum', 0.9),
        weight_decay=best_params.get('weight_decay', 1e-4)
    )
    
    # Learning rate scheduler (more aggressive for EfficientNet)
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='max', patience=4, factor=0.3)
    
    # Training tracking
    train_accs, val_accs, train_losses, val_losses = [], [], [], []
    best_val_acc = 0.0
    patience_counter = 0
    patience = 8  # Reduced patience for EfficientNet
    
    print("ğŸš€ Starting EfficientNet-B3 training...")
    print("=" * 80)
    
    for epoch in range(num_epochs):
        epoch_start = time.time()
        
        # Training phase
        model.train()
        running_loss = 0.0
        correct, total = 0, 0
        
        train_pbar = tqdm(train_loader, desc=f"ğŸƒ Epoch {epoch+1}/{num_epochs}", leave=False)
        for inputs, glcm_feats, labels in train_pbar:
            inputs = inputs.to(device, non_blocking=True)
            glcm_feats = glcm_feats.to(device, non_blocking=True)
            labels = labels.to(device, non_blocking=True)
            
            optimizer.zero_grad()
            outputs = model(inputs, glcm_feats)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            
            running_loss += loss.item()
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
            
            train_pbar.set_postfix({
                'Loss': f'{loss.item():.4f}',
                'Acc': f'{correct/total:.4f}'
            })
        
        train_acc = correct / total
        train_loss = running_loss / len(train_loader)
        train_accs.append(train_acc)
        train_losses.append(train_loss)
        
        # Validation phase
        model.eval()
        val_correct, val_total = 0, 0
        val_loss = 0.0
        
        with torch.no_grad():
            for inputs, glcm_feats, labels in val_loader:
                inputs = inputs.to(device, non_blocking=True)
                glcm_feats = glcm_feats.to(device, non_blocking=True)
                labels = labels.to(device, non_blocking=True)
                
                outputs = model(inputs, glcm_feats)
                loss = criterion(outputs, labels)
                
                val_loss += loss.item()
                _, predicted = torch.max(outputs, 1)
                val_total += labels.size(0)
                val_correct += (predicted == labels).sum().item()
        
        val_acc = val_correct / val_total
        val_loss = val_loss / len(val_loader)
        val_accs.append(val_acc)
        val_losses.append(val_loss)
        
        # Learning rate scheduling
        scheduler.step(val_acc)
        
        epoch_time = time.time() - epoch_start
        
        print(f"â±ï¸  Epoch {epoch+1:2d} ({epoch_time:5.1f}s) | "
              f"Train: {train_acc:.4f} | Val: {val_acc:.4f} | "
              f"LR: {optimizer.param_groups[0]['lr']:.6f}")
        
        # Save best model
        if val_acc > best_val_acc:
            best_val_acc = val_acc
            patience_counter = 0
            torch.save(model.state_dict(), 'best_efficientnet_b3_glcm_model.pth')
            print(f"ğŸ¯ New best validation accuracy: {val_acc:.4f} - Model saved!")
        else:
            patience_counter += 1
            
        if patience_counter >= patience:
            print(f"ğŸ›‘ Early stopping after {epoch+1} epochs")
            break
    
    # Load best model
    model.load_state_dict(torch.load('best_efficientnet_b3_glcm_model.pth'))
    
    # Final evaluation on test set
    print("\nğŸ§ª Final evaluation on test set...")
    model.eval()
    all_preds = []
    all_labels = []
    
    with torch.no_grad():
        for inputs, glcm_feats, labels in tqdm(test_loader, desc="Testing"):
            inputs = inputs.to(device, non_blocking=True)
            glcm_feats = glcm_feats.to(device, non_blocking=True)
            
            outputs = model(inputs, glcm_feats)
            _, predicted = torch.max(outputs, 1)
            
            all_preds.extend(predicted.cpu().numpy())
            all_labels.extend(labels.numpy())
    
    test_accuracy = accuracy_score(all_labels, all_preds)
    
    print(f"ğŸ¯ Final Test Accuracy: {test_accuracy:.4f}")
    print("\nğŸ“Š Classification Report:")
    print(classification_report(all_labels, all_preds, target_names=class_names))
    
    # Plot results
    plot_training_results(train_accs, val_accs, train_losses, val_losses, test_accuracy, class_names, all_labels, all_preds)
    
    return model, test_accuracy, best_val_acc

def plot_training_results(train_accs, val_accs, train_losses, val_losses, test_accuracy, class_names, all_labels, all_preds):
    """Plot training curves and confusion matrix"""
    
    plt.figure(figsize=(15, 5))
    
    # Loss curves
    plt.subplot(1, 3, 1)
    plt.plot(train_losses, label='Train Loss', color='blue')
    plt.plot(val_losses, label='Val Loss', color='red')
    plt.title('EfficientNet-B3 Loss Curves')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    
    # Accuracy curves
    plt.subplot(1, 3, 2)
    plt.plot(train_accs, label='Train Acc', color='blue')
    plt.plot(val_accs, label='Val Acc', color='red')
    plt.title('EfficientNet-B3 Accuracy Curves')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True)
    
    # Validation accuracy progress
    plt.subplot(1, 3, 3)
    plt.plot([acc * 100 for acc in val_accs], 'g-', linewidth=2)
    plt.axhline(y=test_accuracy*100, color='r', linestyle='--', label=f'Test Acc: {test_accuracy:.3f}')
    plt.title('EfficientNet-B3 Validation Accuracy Progress')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy (%)')
    plt.legend()
    plt.grid(True)
    
    plt.tight_layout()
    plt.show()
    
    # Confusion Matrix
    cm = confusion_matrix(all_labels, all_preds)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=class_names, yticklabels=class_names)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title(f'EfficientNet-B3 + GLCM Confusion Matrix - Test Accuracy: {test_accuracy:.4f}')
    plt.tight_layout()
    plt.show()

# Main execution
if __name__ == "__main__":
    print("ğŸš€ EfficientNet-B3 + GLCM with Optuna Optimization for Folder-based DR Dataset")
    print("=" * 80)
    
    # Configuration
    data_dir = '/kaggle/input/diabetic-retinopathy-224x224-2019-data/colored_images'
    batch_size = 12  # Smaller batch size due to larger images (300x300)
    n_trials = 10  # Adjust based on time constraints
    
    print(f"ğŸ“ Dataset directory: {data_dir}")
    print(f"ğŸ”§ Optuna trials: {n_trials}")
    print(f"ğŸ“ Image size: 300x300 (EfficientNet-B3 optimal size)")
    
    # Install efficientnet_pytorch if not available
    if not EFFICIENTNET_AVAILABLE:
        print("\nâš ï¸  Installing efficientnet_pytorch...")
        try:
            import subprocess
            subprocess.check_call(["pip", "install", "efficientnet_pytorch"])
            from efficientnet_pytorch import EfficientNet
            EFFICIENTNET_AVAILABLE = True
            print("âœ… efficientnet_pytorch installed successfully!")
        except:
            print("âŒ Could not install efficientnet_pytorch, using torchvision fallback")
    
    # Create data loaders
    print("\nğŸ—ï¸  Creating EfficientNet-B3 data loaders...")
    start_time = time.time()
    train_loader, val_loader, test_loader, class_names = create_data_loaders(
        data_dir, batch_size=batch_size, val_split=0.2, test_split=0.1
    )
    loader_time = time.time() - start_time
    print(f"â±ï¸  Data loaders created in {loader_time:.2f}s")
    
    n_classes = len(class_names)
    print(f"ğŸ“Š Dataset summary:")
    print(f"   Classes: {class_names}")
    print(f"   Number of classes: {n_classes}")
    
    # Run Optuna optimization
    print("\nğŸ”§ Running Optuna hyperparameter optimization for EfficientNet-B3...")
    best_params, best_score = run_optuna_optimization(
        train_loader, val_loader, n_classes, n_trials=n_trials
    )
    
    # Train final model
    print("\nğŸ‹ï¸  Training final EfficientNet-B3 model with optimized hyperparameters...")
    final_model, test_accuracy, best_val_acc = train_final_efficientnet_b3_model(
        train_loader, val_loader, test_loader, class_names, best_params, num_epochs=40
    )
    
    # Final summary
    print(f"\nğŸ‰ EFFICIENTNET-B3 + GLCM FINAL RESULTS:")
    print("=" * 60)
    print(f"ğŸ”§ Optuna Trials: {n_trials}")
    print(f"ğŸ¯ Best Validation Accuracy: {best_val_acc:.4f} ({best_val_acc*100:.2f}%)")
    print(f"ğŸ§ª Final Test Accuracy: {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")
    print(f"ğŸ“Š Classes: {class_names}")
    print(f"ğŸ—ï¸  Architecture: EfficientNet-B3 (1536 features) + GLCM (12 features)")
    print(f"ğŸ“ Input Size: 300x300 pixels")
    print("=" * 60)
    
    print("\nğŸ”§ Optimized Hyperparameters:")
    for param, value in best_params.items():
        print(f"   {param}: {value}")
    
    print("\nâœ… EfficientNet-B3 training completed! Model saved as 'best_efficientnet_b3_glcm_model.pth'")
    
    # Model summary
    total_params = sum(p.numel() for p in final_model.parameters())
    trainable_params = sum(p.numel() for p in final_model.parameters() if p.requires_grad)
    print(f"\nğŸ“ˆ Model Statistics:")
    print(f"   Total parameters: {total_params:,}")
    print(f"   Trainable parameters: {trainable_params:,}")
    print(f"   Model size: ~{total_params * 4 / (1024**2):.1f} MB")
    
    # Performance comparison tip
    print(f"\nğŸ’¡ Performance Tips:")
    print(f"   â€¢ EfficientNet-B3 uses 300x300 input (vs 224x224 for ResNet)")
    print(f"   â€¢ Lower learning rates work better (5e-6 to 5e-3)")
    print(f"   â€¢ Weight decay helps prevent overfitting")
    print(f"   â€¢ Smaller batch sizes due to memory requirements")
```
